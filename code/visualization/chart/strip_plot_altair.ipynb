```python
import altair as alt
import pandas as pd
import numpy as np
import os
```

```python
alt.data_transformers.disable_max_rows()
```

```python
def hms_to_sec(hms):
    h = int(hms.split(':')[0])
    m = int(hms.split(':')[1])
    s = int(hms.split(':')[2])

    if h < 0:
        raise Exception('not available number for h')

    if (len(hms.split(':')[1]) != 2) or (m < 0) or (m > 60):
        raise Exception('not available number for m')

    if (len(hms.split(':')[2]) != 2) or (s < 0) or (s > 60):
        raise Exception('not available number for s')

    sec = (h * 3600) + (m * 60) + s
    return sec
```

```python
root = ''
path_gtfolder = os.path.join(root,'gt')
path_predfolder = os.path.join(root, 'pred_winsize0')

list_gtfiles = os.listdir(path_gtfolder)
list_predfiles = os.listdir(path_predfolder)
```

```python
# select file, tools
file_name = ''
tools = ''

# get predfile
for pred_file_ele in list_predfiles:
    if pred_file_ele[:24] == file_name[:24]:
        predfile = pred_file_ele
    
# get gtfile
for gt_file_ele in list_gtfiles:
    if gt_file_ele[:24] == file_name[:24]:
        gtfile = gt_file_ele    

path_predfile = os.path.join(path_predfolder, predfile)
path_gt = os.path.join(path_gtfolder, gtfile)

df_pred = pd.read_csv(path_predfile)
df_gt = pd.read_csv(path_gt)

df_pred_tool = df_pred[tools]
df_gt_tool = df_gt[tools]
```

```python
list_timeline_gt = []
list_timeline_notoverlap = []
list_timeline_pred = []

# gt table
for idx, appear_ in enumerate(df_gt_tool):
    if appear_ == 1:
        list_timeline_gt.append(idx)
        list_name_gt = [' GT']*len(list_timeline_gt)

# gt, pred overlap table
for idx, appear_ in enumerate(df_gt_tool):
    if appear_ != df_pred_tool[idx]:
        list_timeline_notoverlap.append(idx)
        list_name_notoverlap = ['False']*len(list_timeline_notoverlap)
        
# pred table
for idx, appear_ in enumerate(df_pred_tool):
    if appear_ == 1:
        list_timeline_pred.append(idx)
        list_name_pred = [' pred']*len(list_timeline_pred)   
        
list_timeline = list_timeline_gt + list_timeline_notoverlap + list_timeline_pred
list_name = list_name_gt + list_name_notoverlap + list_name_pred
```

```python
display(len(list_timeline))
display(len(list_name))
```

```python
df_totick = pd.DataFrame()
df_totick['time'] = list_timeline
df_totick['name'] = list_name
```

```python
alt.Chart(df_totick).mark_tick().encode(
    x='time:Q',
    y='name:O',
    color = 'name'
    
).properties(
    width=1500,
    height=200
)
```
